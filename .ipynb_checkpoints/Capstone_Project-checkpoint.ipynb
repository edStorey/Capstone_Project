```python
from Classes.Image_Reader import image_Reader
from Classes.Full_Image_Scan import image_Scan
from Classes.Digit_Model import digitModel
from Classes.Full_Number_Model import fullNumberModel
import matplotlib.pyplot as plt
import numpy as np

# Initialise file paths as variables
tain_image_Path_32x32 = '../Capstone_Project/train_32x32/'
train_csv_path_32x32  = "CSV_Data/train_Image_Digits.txt"
test_image_Path_32x32 = '../Capstone_Project/test_32x32/'
test_csv_path_32x32  = "CSV_Data/test_Image_Digits.txt"
classes_32x32 = 10
width = 32
height = 32


# image_Reader loads image data into and array and processes the validation data
train_32x32 = image_Reader(tain_image_Path_32x32, width, height, train_csv_path_32x32, classes_32x32)

test_32x32 = image_Reader(test_image_Path_32x32, width, height, test_csv_path_32x32, classes_32x32)
```

```python
# Train the inital neural network, identifying single 32x32 images
model_32x32 = digitModel(train_32x32.X_norm, train_32x32.one_Hot(), test_32x32.X_norm, test_32x32.one_Hot(), train_32x32.data, height, width, classes_32x32)
```

```python
print(len(test_32x32.X_norm), len(test_32x32.one_Hot()))
#print(model_32x32.predictions[len(model_32x32.predictions) - 20:len(model_32x32.predictions)])

# Show model summary
model_32x32.model.summary()
```

```python
# Plot training accuracy and Loss
lot accuracy and loss over time
epochs= range(1, len(model_32x32.history.history['loss'])+1)
plt.plot(epochs, np.array(model_32x32.history.history['acc']), 'b', label='Training acc')
plt.title('Training accuracy')
plt.show()
plt.plot(epochs, model_32x32.history.history['loss'], 'b', label='Validation acc')
plt.title('Training Loss')
plt.show()
```

```python
train_Image_Path = 'train/'
train_CSV_Path = 'CSV_Data/train_images_Full_Numbers.txt'
test_Image_Path = 'test/'
test_CSV_Path = 'CSV_Data/test_images_Full_Numbers.txt'
classes_Full = 20
full_height = 64
full_width = 96

# Load full image data and resize  to 64,96 pixels
train= image_Reader(train_Image_Path, full_width, full_height, train_CSV_Path, classes_Full)

test= image_Reader(test_Image_Path, full_width, full_height, test_CSV_Path, classes_Full)
```

```python
model32x32_Path = 'model_keras.h5'
model32x32_weights = 'model_weights_32x32.h5'
size32x32 = 32

# Scan through each full image with 32x32 windows and output the likeliness of each digit existing in each window
train_predictions = image_Scan(train.X, model32x32_Path, model32x32_weights, full_width, full_height, size32x32, classes_32x32)

test_predictions = image_Scan(test.X, model32x32_Path, model32x32_weights, full_width, full_height, size32x32, classes_32x32)
```

```python
# Enter window likeliness for each digit into a neural network
scanned_Model_Data = fullNumberModel(train_predictions.predictions, train.binary_One_Hot(),
                                     test_predictions.predictions, test.binary_One_Hot(),
                                     train_predictions.array_size, classes_Full)
```

```python
# shows summary of model
scanned_Model_Data.model.summary()
```

```python
# Output overall algorithm success
print(scanned_Model_Data.performance.mean())
```
